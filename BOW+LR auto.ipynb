```python
pasta = 'treinamento/positivo'
arquivos = os.listdir(pasta)
train = []
rotulos = []
# Itera pelos arquivos e lรช o conteรบdo de cada um
for arquivo in arquivos:
    # Verifica se รฉ um arquivo e se รฉ um arquivo txt
    if os.path.isfile(os.path.join(pasta, arquivo)) :
        if (os.path.join(pasta, arquivo)[-3:] == "txt"):
            with open(os.path.join(pasta, arquivo), 'r',encoding='utf-8') as f:
                conteudo = f.read()
        elif(os.path.join(pasta, arquivo)[-3:] == "xml"):
            tree = etree.parse(os.path.join(pasta, arquivo))
            root = tree.getroot()
            conteudo = BeautifulSoup(root.xpath('//xml//body//Texto')[0].text, "html.parser").get_text()
        train.append(remover_caracteres(conteudo))
rotulos.append([1 for i in range(len(train))])
```

```python
pasta = 'treinamento/negativo'
arquivos = os.listdir(pasta)
arquivosLidos=0
# Itera pelos arquivos e lรช o conteรบdo de cada um
for arquivo in arquivos:
    # Verifica se รฉ um arquivo e se รฉ um arquivo txt
    if os.path.isfile(os.path.join(pasta, arquivo)):
        if (os.path.join(pasta, arquivo)[-3:] == "txt"):
            with open(os.path.join(pasta, arquivo), 'r',encoding='utf-8') as f:
                conteudo = f.read()
        elif (os.path.join(pasta, arquivo)[-3:] == "xml"):
            tree = etree.parse(os.path.join(pasta, arquivo))
            root = tree.getroot()
            conteudo = BeautifulSoup(root.xpath('//xml//body//Texto')[0].text, "html.parser").get_text()
        train.append(remover_caracteres(conteudo))
    arquivosLidos = arquivosLidos + 1
rotulos.append([0 for i in range(arquivosLidos)])
rotulos = rotulos[0]+rotulos[1]
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(train)
```

```python
clf = LogisticRegression()
clf.fit(X, rotulos)
```

```python
def avaliar(texto):
    novo_vetor = vectorizer.transform([texto])
    predicao = clf.predict(novo_vetor)
    if predicao[0] == 1:
        return True
    else:
        return False
```
